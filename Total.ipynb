```python
import pandas as pd
import openpyxl



# Ouvrez le fichier Excel avec openpyxl
wb = openpyxl.load_workbook('Total_Question.xlsx')

# Obtenez le nom du premier onglet
premier_onglet = wb.sheetnames[0]

# Fermez le fichier Excel
wb.close()

# Lisez le premier onglet avec pandas
df = pd.read_excel('Total_Question.xlsx', sheet_name=premier_onglet)

# Maintenant, 'df' contient les données du premier onglet en excluant les 7 premières lignes
onglet_a_traiter = premier_onglet
```

```python
df
```

## Cycle

```python

# Initialise une variable pour stocker la valeur du cycle
cycle_actuel = None
nouvelle_colonne_cycle = []

# Crée la nouvelle colonne "Cycle" en se basant sur les colonnes 1 et 3
for index, row in df.iterrows():
    if row[2] == 0:
        # Si la colonne 9 contient un 0, met la valeur de la colonne 1 dans "Cycle"
        cycle_actuel = row[0]
    nouvelle_colonne_cycle.append(cycle_actuel)

# Ajoute la nouvelle colonne "Cycle" au DataFrame
df['Cycle'] = nouvelle_colonne_cycle

```

```python
df
```

## Sous-Catégories

```python

# Initialise une variable pour stocker la valeur de la sous-catégorie
sous_categories_actuelles = None
nouvelle_colonne_sous_categories = []

# Parcours les lignes du DataFrame
for index, row in df.iterrows():
    if  pd.isna(row[1]) and pd.isna(row[2]) and pd.isna(row[3]):
        # Si la colonne 6 est remplie et les colonnes 8 et 9 sont vides,
        # met la valeur de la colonne 2 dans "Sous-Catégories"
        sous_categories_actuelles = row[0]
    nouvelle_colonne_sous_categories.append(sous_categories_actuelles)

# Ajoute la nouvelle colonne "Sous-Catégories" au DataFrame
df['Sous-Catégories'] = nouvelle_colonne_sous_categories

df
```

```python
# Supprime la colonne 3 (en comptant à partir de 0)
df = df.drop(df.columns[2], axis=1)
```

```python
df = df.dropna(subset=[df.columns[1]])
```

```python
df
```

```python
# Créez une nouvelle colonne 'N° Question' avec des valeurs de 'Q-0001' en incrémentant
df.insert(0, 'N° Question', ['Q-{:04d}'.format(i) for i in range(1, len(df) + 1)])

# Assurez-vous que la nouvelle colonne a été ajoutée en première position
```

```python
# Réindexe le DataFrame
df = df.reset_index(drop=True)
```
